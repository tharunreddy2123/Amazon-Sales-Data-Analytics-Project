```python
#1. Which regions have the highest total sales revenue?
Highest_total_sales_revenue = data.groupby('Region')['Total Revenue'].sum().sort_values(ascending=False)
Highest_total_sales_revenue.head(1)
```

```python
plt.figure(figsize=(10,5))
Highest_total_sales_revenue.plot(kind='line', marker='o', title='Highest Total Sales Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
#2- What is the average unit price and unit cost for each item type?
average_unit_price_cost = data.groupby('Item Type')[['Unit Price', 'Unit Cost']].mean()
average_unit_price_cost
```

```python
#3- Which country has the highest total profit?
highest_total_profit_country = data.groupby('Country')['Total Profit'].sum().sort_values(ascending=False)
print(highest_total_profit_country.head(1))
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x=highest_total_profit_country.index, y=highest_total_profit_country.values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Highest Total Profits by Country')
plt.show()
```

```python
Sales_Channel_Order_Priority_Distribution= data.groupby(data['Sales Channel']) ['Order Priority'].value_counts()
Sales_Channel_Order_Priority_Distribution
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Sales_Channel_Order_Priority_Distribution.index.get_level_values(0), 
            y=Sales_Channel_Order_Priority_Distribution.values, 
            hue=Sales_Channel_Order_Priority_Distribution.index.get_level_values(1))
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.title('Sales Channel Order Priority Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
#5- What is the average order processing time (duration between order and ship dates) for each sales channel?
data['Order Processing Time'] = (data['Ship Date'] - data['Order Date']).dt.days
average_processing_time = data.groupby('Sales Channel')['Order Processing Time'].mean()
average_processing_time
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(x=average_processing_time.index, y=average_processing_time.values)
plt.xlabel('Sales Channel')
plt.ylabel('Average Order Processing Time (days)')
plt.title('Average Order Processing Time by Sales Channel')
plt.yticks(np.arange(0,25,1))

plt.show()
```

```python
#6- Which item types have the highest and lowest total sales?
highest_total_sales = data.groupby('Item Type')['Total Revenue'].sum().sort_values(ascending=False)
highest_total_sales.head(1)
```

```python
lowest_total_sales = data.groupby('Item Type')['Total Revenue'].sum().sort_values(ascending=True)
lowest_total_sales.head(1)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Unit Price', y='Total Revenue', hue='Item Type')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.title('Scatter Plot of Unit Price vs Total Revenue by Item Type')
plt.show()
```

```python
#7- How does the order priority vary across different regions?

order_priority_by_region = data.groupby('Region')['Order Priority'].value_counts().unstack().fillna(0)
order_priority_by_region
```

```python
order_priority_by_region.plot(kind='bar', figsize=(12, 8))
plt.title('Order Priority Distribution Across Different Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Order Priority')
plt.show()
```

```python
#8- What is the correlation between unit price and total profit?
correlation = data['Unit Price'].corr(data['Total Profit'])
correlation
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Unit Price', y='Total Revenue', hue='Item Type')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.title('Scatter Plot of Unit Price vs Total Revenue by Item Type')
plt.show()
```

```python
#9- Are there any seasonal trends or patterns in the sales data?

data['Order Month'] = data['Order Date'].dt.month
monthly_sales = data.groupby('Order Month')['Total Revenue'].sum()
monthly_sales
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x=monthly_sales.index, y=monthly_sales.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.title('Highest Total Profits by Country')
plt.show()
```

```python
#10- How does the number of units sold vary across different countries?

Diff_countries_by_unit_sold= data.groupby(data['Country'])['Units Sold'].sum().reset_index(name= 'Unit Sold')
pd.set_option('display.max_rows',None)
Diff_countries_by_unit_sold
```

```python
plt.figure(figsize= (18,7))
sns.barplot( data= Diff_countries_by_unit_sold, x= 'Country', y= 'Unit Sold')
plt.xticks(rotation= 90)
plt.yticks(np.arange(0,26000,1000))
plt.show()
```

```python
sales_revenue_by_countries=  data.groupby(data['Country']) ['Total Revenue'].sum().reset_index(name= 'Total Revenue')
sales_revenue_by_countries
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x= sales_revenue_by_countries['Country'], y= sales_revenue_by_countries['Total Revenue'])
plt.xticks(rotation= 90)
plt.show()
```

```python
#12- What is the distribution of unit prices for each item type?
unit_price_and_item_type_distribution= data.groupby(data['Item Type'])['Unit Price'].sum().reset_index(name= 'Unit Price')
unit_price_and_item_type_distribution
```

```python
plt.pie(x= unit_price_and_item_type_distribution['Unit Price'], labels= unit_price_and_item_type_distribution['Item Type'],autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
13- Which sales channel has the highest average unit price?

Highest_avg_unit_price_for_sales_channel= data.groupby(data['Sales Channel']) ['Unit Price'].mean().reset_index(name= 'new')
Highest_avg_unit_price_for_sales_channel
```

```python
plt.pie(x= Highest_avg_unit_price_for_sales_channel['new'],labels=Highest_avg_unit_price_for_sales_channel['Sales Channel'], autopct= '%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#14- Are there any outliers in the total cost distribution?
q1= data['Total Cost'].quantile(0.25)
q3= data['Total Cost'].quantile(0.75)

iqr= q3-q1

lower_fence= q1-1.5*iqr
upper_fence= q3+1.5*iqr

outliers= data[(data['Total Cost']<lower_fence)| (data['Total Cost']>upper_fence)].reset_index(drop= True)
outliers
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='Total Cost')
plt.title('Boxplot of Total Cost')
plt.ylabel('Total Cost')
plt.show()
```
